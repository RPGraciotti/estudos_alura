# Aula 2

### Desafio: 
- remover os dados em listas dentro do DataFrame;
- verificar os tipos de dados;
- identificar colunas numéricas;
- transformar a coluna numérica para o tipo numérico. 

````python
vendas = vendas.explode(list(vendas.columns)[1:])
vendas.reset_index(inplace = True, drop = True)
vendas
````

````python
vendas.info()
````

````python
vendas['Valor da compra'] = vendas['Valor da compra'].apply(lambda x: x.replace('R$', '').replace(',','.').strip())
````

````python
vendas['Valor da compra'] = vendas['Valor da compra'].astype(np.float64)
````

````python
vendas
````

````python
locacao = locacao.explode(list(locacao.columns)[1:])
locacao.reset_index(inplace = True, drop = True)
locacao
````

````python
locacao['valor_aluguel'] = locacao['valor_aluguel'].apply(lambda x: x.replace('$', '').replace(',0','').replace('reais','').strip())
````

````python
locacao['valor_aluguel'] = locacao['valor_aluguel'].astype(np.int64)
````

````python
locacao
````

# Aula 3

### Desafio:
- manipular coluna cliente para que o resultado tenha apenas letras minúsculas com a ausência de caracteres especiais ou números

````python
vendas['Cliente'] = vendas['Cliente'].str.replace('[^a-zA-Z\s]','', regex = True).str.lower()
````

````python
vendas
````

### Desafio:
- manipular os textos na coluna apartamento para remover o texto (blocoAP) do dataframe

````python
locacao['apartamento'] = locacao['apartamento'].str.replace('\(blocoAP\)','', regex = True)
````

em regex, é necessário usar a \ sempre que passar um caractere especial, de forma a dizer "procure literalmente por esse caractere".

````python
locacao['apartamento']
````

- transformar os dados da coluna Data de venda em datetime e busque uma forma de visualização de subconjunto que possa contribuir no objetivo do contexto que os dados estão inseridos:

```
O objetivo desse projeto é realizar uma análise dos resultados de um evento com os clientes de uma empresa de vendas online. Foi coletado um conjunto de dados que contém os clientes que mais gastaram com produtos dentro de 5 dias de vendas, que é o período de duração do evento. ### Desafio:
```
- Transforme os dados das colunas datas_de_pagamento e datas_combindas_pagamento do dataframe locacao para o tipo datetime e busque uma forma de visualização de subconjunto que possoa contribuir no objetivo do contexto que os dados estão inseridos:

```A administração de condomínios é uma tarefa que requer muita atenção e organização. Entre as diversas responsabilidades da gestão está o recebimento dos aluguéis dos locatários. Para garantir uma boa saúde financeira do empreendimento, é fundamental que esses pagamentos sejam feitos de forma regular e pontual. No entanto, sabemos que nem sempre isso acontece. Pensando nisso, proponho um desafio de tratamento de dados com o objetivo de analisar o atraso no pagamento de aluguéis no condomínio fictício de alguns moradores.```

````python
locacao[['datas_de_pagamento','datas_combinadas_pagamento']]
````

````python
locacao['datas_combinadas_pagamento'] = pd.to_datetime(locacao['datas_combinadas_pagamento'], format = '%d/%m/%Y')
locacao['datas_de_pagamento'] = pd.to_datetime(locacao['datas_de_pagamento'], format = '%d/%m/%Y')
````

````python
locacao[['datas_combinadas_pagamento', 'datas_de_pagamento']]
````

````python
locacao['atraso'] = locacao['datas_de_pagamento'] - locacao['datas_combinadas_pagamento']
````

````python
locacao.groupby('apartamento')['atraso'].mean().sort_values(ascending = False)
````
