# Projeto citrus

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
citrus = np.loadtxt('https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv', delimiter = ',', usecols = np.arange(1,6,1), skiprows = 1)

citrus

# nesse caso, o usecols e skiprows estão sendo usados porque tanto a primeira linha quanto a primeira coluna tem strings, e não existe uma função "skipcols"
```

```python
diametro_laranja = citrus[:5000, 0]
diametro_toranja = citrus[5000:, 0]
peso_laranja = citrus[:5000, 1]
peso_toranja = citrus[5000:, 1]
```

```python
plt.plot(diametro_toranja, peso_toranja)
plt.plot(diametro_laranja, peso_laranja)
plt.legend(['Toranja', 'Laranja'])
```

Coeficiente angular laranjas:

X = diametro

Y = peso

```python
n = np.size(diametro_laranja)
X_laranja = diametro_laranja
Y_laranja = peso_laranja
```

```python
a_laranja = (n * np.sum(X_laranja*Y_laranja) - np.sum(X_laranja) * np.sum(Y_laranja)) / (n * np.sum(X_laranja**2) - np.sum(X_laranja)**2)

b_laranja = np.mean(Y_laranja) - a_laranja * np.mean(X_laranja)

print(a_laranja, b_laranja)
```

```python
y_laranja = a_laranja * X_laranja + b_laranja
y_laranja
```

```python
plt.plot(diametro_laranja, peso_laranja)
plt.plot(diametro_laranja, y_laranja)
```

Coeficiente angular toranjas:

X = diametro

Y = peso

```python
n = np.size(diametro_toranja)
X_toranja = diametro_toranja
Y_toranja = peso_toranja
```

```python
a_toranja = (n * np.sum(X_toranja*Y_toranja) - np.sum(X_toranja) * np.sum(Y_toranja)) / (n * np.sum(X_toranja**2) - np.sum(X_toranja)**2)

b_toranja = np.mean(Y_toranja) - a_toranja * np.mean(X_toranja)

print(a_toranja, b_toranja)
```

```python
y_toranja = a_toranja * X_toranja + b_toranja
y_toranja
```

```python
plt.plot(diametro_toranja, peso_toranja)
plt.plot(diametro_toranja, y_toranja)
```

Coeficiente angular usando números aleatórios

Laranjas

```python
np.random.seed(84)
coef_angulares_laranja = np.random.uniform(low = 0, high = 30, size = 100)

X_laranja = diametro_laranja
Y_laranja = peso_laranja

norma_laranja = np.array([])

for i in range(100):
    norma_laranja = np.append(norma_laranja, np.linalg.norm(Y_laranja - (coef_angulares_laranja[i] * X_laranja + 17)))

print(np.min(norma_laranja))

print(coef_angulares_laranja[norma_laranja == np.min(norma_laranja)])
```

Toranjas

```python
np.random.seed(84)
coef_angulares_toranja = np.random.uniform(low = 0, high = 30, size = 100)

X_toranja = diametro_toranja
Y_toranja = peso_toranja

norma_toranja = np.array([])

for i in range(100):
    norma_toranja = np.append(norma_toranja, np.linalg.norm(Y_toranja - (coef_angulares_toranja[i] * X_toranja + 17)))

print(np.min(norma_toranja))

print(coef_angulares_toranja[norma_toranja == np.min(norma_toranja)])
```
