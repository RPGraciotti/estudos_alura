# Data Visualization 2

## Conhecendo a biblioteca Matplotlib

```python
import pandas as pd
```

```python
df = pd.read_table('./data/imigrantes_canada.txt', sep = ',')
```

```python
df.head()
```

```python
df.info()
```

```python
df.set_index('País', inplace = True)
df.head()
```

```python
anos = list(map(str, range(1980, 2014)))
anos
# Isso aqui vai fazer sentido porque eu vou ficar extraindo dados de outros países mais pra frente
```

```python
brasil = df.loc['Brasil', anos]
brasil
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
```

```python
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Nº de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show()
```

### Desafio: Adicionar uma linha com os dados da Argentina

```python
arg = df.loc['Argentina', anos]
arg_dict = {'ano': arg.index.tolist(), 'imigrantes': arg.values.tolist()}
dados_arg = pd.DataFrame(arg_dict)
```

```python
plt.figure(figsize = (8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'], label = 'Brasil')
plt.plot(dados_arg['ano'], dados_arg['imigrantes'], label = 'Argentina')
plt.title('Imigração do Brasil e Argentina para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Nº de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```

1) Há alguma tendência ou padrão comum nos dados dos dois países?
R: Sim, o padrão é bastante similar, diferindo que o pico da imigração Argentina foi um pouco menor e alguns anos antes que o Brasil.

2) Quais são os períodos com maior número de imigrantes nos dois países?
R: O pico da imigração Aegentina ocorre no começo da década de 2000, enquanto no Brasil o pico é na segunda metade da década.

```python
fig, axs = plt.subplots(2, 2, figsize = (10,6))
fig.subplots_adjust(hspace = 0.5, wspace = 0.3) #hpsace = height space, wspace = width space
fig.suptitle('Imigração dos quatro maiores países da América do Sul para o Canadá\n1980 a 2013')

axs[0, 0].plot(df.loc['Brasil', anos])
axs[0, 0].set_title('Brasil')

axs[0, 1].plot(df.loc['Colômbia', anos])
axs[0, 1].set_title('Colômbia')

axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

axs[1, 1].plot(df.loc['Peru', anos])
axs[1, 1].set_title('Peru')

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel('Ano')
    ax.set_ylabel('Imigrantes')

ymin = 0
ymax = 7000 # deve dar pra extrair a partir dos próprios dados

for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

plt.show()
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
df_vendas = pd.DataFrame(vendas_2022, index = lojas)
meses = list(map(str, ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']))
```

```python
fig, axs = plt.subplots(2, 2, figsize = (10,6))
fig.subplots_adjust(hspace = 0.5, wspace = 0.3) #hpsace = height space, wspace = width space
fig.suptitle('Vendas das lojas ABCD em 2022')

axs[0, 0].plot(df_vendas.loc['A', meses])
axs[0, 0].set_title('Loja A')

axs[0, 1].plot(df_vendas.loc['B', meses])
axs[0, 1].set_title('Loja B')

axs[1, 0].plot(df_vendas.loc['C', meses])
axs[1, 0].set_title('Loja C')

axs[1, 1].plot(df_vendas.loc['D', meses])
axs[1, 1].set_title('Loja D')

for ax in axs.flat:
    ax.set_xlabel('Mês')
    ax.set_ylabel('Vendas')

ymin = 0
ymax = df_vendas.max().max() + 50

for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

plt.show()
```

```python
with plt.style.context('ggplot'):
  fig, ax = plt.subplots(figsize=(8, 4))
  ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3)
  ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013', fontsize=20, loc='left')
  ax.set_ylabel('Número de imigrantes', fontsize=14)
  ax.set_xlabel('Ano', fontsize=14)
  ax.yaxis.set_tick_params(labelsize=12)
  ax.xaxis.set_tick_params(labelsize=12)
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  plt.show()
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal',  'tomato']
fig, ax = plt.subplots(figsize = (12, 5))
ax.barh(america_sul.index, america_sul['Total'], color = cores)
ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', fontsize = 18, loc = 'left')
ax.set_xlabel('Número de de Imigirantes', fontsize = 14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)
plt.show()
```

```python
america_sul_sorted = america_sul.sort_values('Total', ascending = True)
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal',  'tomato']
fig, ax = plt.subplots(figsize = (12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color = cores)
ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', fontsize = 18, loc = 'left')
ax.set_xlabel('Número de de Imigirantes', fontsize = 14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)
plt.show()
```

```python
cores = []
for pais in america_sul_sorted.index:
    if pais == 'Brasil': 
        cores.append('green')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize = (12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color = cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o canadá de 1980 a 2013', fontsize = 18, loc = 'left')
ax.set_xlabel('Número de de Imigirantes', fontsize = 14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)
plt.show()
```

```python
cores = []
for pais in america_sul_sorted.index:
    if pais == 'Brasil': 
        cores.append('green')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize = (12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color = cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o canadá de 1980 a 2013', fontsize = 18, loc = 'left')
ax.set_xlabel('Número de de Imigirantes', fontsize = 14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)

for i, v in enumerate(america_sul_sorted['Total']):
    ax.text(v + 20, i, str(v), color = 'k', fontsize = 10, ha = 'left', va = 'center') # onde o texto vai adicionado em cada barra, ou seja, 20 pontos depois para dar espaço

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis = 'both', which = 'both', length = 0)

fig.savefig('imigracao_america_sul.png', transparent = False, dpi = 300, bbox_inches = 'tight')

plt.show()
```
